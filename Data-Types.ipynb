# Data Type Conversion

```python
a=10 #Integer
b = str(a) #convert to string
type(b)
```

```python
a=10.64 #float
b=int(a) #convert to integer
type(b)
```

```python
a='Hello' #String
b=list(a) #convert to list
print(b)
type(b)
```

```python
a=[(1,'a'),(2,'b'),(3,'c'),(4,'d')]  #List
b=dict(a)
print(b)
type(b)
```

```python
a=[[1,'a'],[2,'b'],[3,'c'],[4,'d']]  #Same we can do in the form of nested list.
b=dict(a)
print(b)
type(b)
```

```python
#Tuples
a=((1,'a'),(2,'b'),(3,'c'),(4,'d'))  #Same we can do in the form of nested tuple.
b=dict(a)
print(b)
type(b)
```

```python
a= True #boolean
b=float(a)  #convert to float
print(b)
type(b)
```

```python
a = 67  #integer
b = complex(a) #convert to complex
print(b)
type(b)
```

```python
a={1,2,3,4,5,6}  #set
b=list(a)  #convert to list
print(b)
type(b)
```
