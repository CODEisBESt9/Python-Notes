# Data Structures

# What Are Data Structures?
A way to organize and store data so it can be used effectively.

They help:

Reduce time and space complexity

Enable faster searching, sorting, and updating

Support real-world applications like databases, compilers, and AI models

# Types of Data Structures
Linear Structures

Array: Fixed-size, indexed collection of elements.

Linked List: Dynamic structure with nodes pointing to the next.

Stack: LIFO (Last In First Out) structure—used in undo operations.

Queue: FIFO (First In First Out)—used in scheduling tasks.

Non-Linear Structures

Tree: Hierarchical structure—used in file systems and decision trees.

Graph: Nodes connected by edges—used in networks and recommendation engines.

Heap: Specialized tree for priority-based tasks.

Hash-Based

Hash Table / HashMap: Key-value pairs with fast lookup—used in dictionaries and caches.
