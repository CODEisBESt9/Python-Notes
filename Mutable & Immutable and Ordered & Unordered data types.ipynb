# Bitwise Operators

```python
#First it convert the value into binary number, then only perform binary operations.
a=5 # In binary, 5 = 0000 0101
b = 3  # In binary, 3 = 0000 0011
```

```python
a & b  # 0000 0101 & 0000 0011 = 0000 0001 (1)  AND operation
```

```python
a | b  # 0000 0101 | 0000 0011 = 0000 0111 (7) OR operation
```

```python
a ^ b  # 0000 0101 ^ 0000 0011 = 0000 0110 (6) XOR operation
```

```python
~a  # ~ 0000 0101  = 0000 0101 + 1 = 0000 0110 (-6) 1's Compliment
```

```python
a << b  # a * 2^b = 5 * 2^3 = 40
```

```python
a >> b  # a / 2^b = 5 / 2^3 = 0 
```
